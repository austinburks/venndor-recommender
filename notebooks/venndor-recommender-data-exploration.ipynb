```python
import pandas as pd
import requests
from sklearn.linear_model import LogisticRegression
from IPython.core.display import display, HTML
import os
from dotenv import load_dotenv, find_dotenv
```

### Open the locally stored api key

```python
_ = load_dotenv(find_dotenv())
```

```python
key = os.environ.get("API_KEY")
base_url = os.environ.get("API_URL")
```

### Load in the users and post history tables

```python
base_url = 'http://bitnami-dreamfactory-b106.cloudapp.net/api/v2/venndor_test/_table/'

users = base_url + 'test_users' + '?api_key=%s' % key
posts = base_url + 'test_seenposts' + '?api_key=%s' % key
items = base_url + 'test_items' + '?api_key=%s' % key
```

### Do the url request, converting to JSON

```python
# Generating users JSON
users_response = requests.get(users)
users_json = users_response.json()

# Generating posts JSON
posts_response = requests.get(posts)
posts_json = posts_response.json()

# Generating items JSON
items_response = requests.get(items)
items_json = items_response.json()
```

### Creating the users DF and displaying top 5 results

```python
df_users = pd.DataFrame(
            users_json[list(users_json.keys())[0]]
                        )
display(HTML(df_users.head(5).to_html()))
```

### Creating the posts DF and displaying top 5 results

```python
df_posts = pd.DataFrame(
            posts_json[list(posts_json.keys())[0]]
                        )
display(HTML(df_posts.dropna(axis=0, thresh=60).head(5).to_html()))
```

```python
df_items = pd.DataFrame(
            items_json[list(items_json.keys())[0]]
                        )

display(HTML(df_items.head(5).to_html()))
```

### We need to do some additional manipulation to the seen posts to get it into a usable format

```python
df_posts_melt = pd.melt(df_posts, id_vars='_id', var_name='Item', value_name='Date')
display(HTML(df_posts_melt.dropna().head(10).to_html()))
```

### Now that we have the history of every seen post, it is time to find out whether the person made an offer on that post

```python
id_list = [var for var in df_items['_id']]
test = dict(zip(id_list, df_items['matches']))
print(test)
```
