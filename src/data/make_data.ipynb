```python
import pandas as pd
import requests
import os
from dotenv import load_dotenv, find_dotenv
```

## Loading environment variables

```python
_ = load_dotenv(find_dotenv())
```

```python
API_KEY, API_URL = os.environ.get('API_KEY'), os.environ.get('API_URL')
```

```python
def load_venndor_data(table=None, API_KEY=API_KEY, API_URL=API_URL):
    if table == None:
        raise ValueError('Please supply a valid table name')
    else:
        response = requests.get(API_URL + '%s?api_key=%s'  % (table, API_KEY))
        json_response = response.json()
        return(json_response)
```

```python
for table_name in ['items', 'matches', 'posts', 'seenPosts', 'users']:
    df = load_venndor_data(table='users')
    df = df[list(df.keys())[0]]
    pd.io.json.json_normalize(df).to_csv('%s.csv' % table_name)                    
    print(table_name)
    print()
    print(pd.io.json.json_normalize(df).head(1))
    print()
```
